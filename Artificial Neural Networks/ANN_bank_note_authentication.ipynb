# Artifical neural network for Bank Note Authentication using Keras

### Importing the libraries

```python
import numpy as np
import pandas as pd
import keras
```

## Part 1 - Data Preprocessing





### Importing the dataset

```python
dataset = pd.read_csv('/content/drive/MyDrive/Deep Learning Examples/bank_note_data.txt', header = None )
```

```python
dataset.head(5)
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Part 2 - Building the ANN

### 1. Initializing the ANN

```python
from keras.models import Sequential
ann = Sequential()
```

### 2. Adding the input layer and the first hidden layer

```python
from keras.layers import Dense
ann.add(Dene(4,input_dim=4,activation = 'relu'))
```

### 3. Adding the second hidden layer

```python
ann.add(Dense(8,activation='relu'))
```

### 4. Adding the output layer

```python
ann.add(Dense(1,activation='sigmoid'))
```

## PART 3 - Training the Artificial Neural Network

### Compiling the ANN

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

### Training the ANN on the Training set

```python
ann.fit(X_train,y_train, epochs = 100, verbose = 2)
```

## Part 4 - Making the predictions and evaluating the model

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```
