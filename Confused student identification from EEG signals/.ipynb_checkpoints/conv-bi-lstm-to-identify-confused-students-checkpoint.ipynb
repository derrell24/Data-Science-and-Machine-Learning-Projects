## Motivation
Poor performance in the learning process mainly stems from confusion. Detecting confusion in a human’s mind in real time is essential for online education to improve the experience. Massive Open  Online Courses  (MOOC) serve  millions  of  students  at  the  same  time,  but  has several shortcomings as compared to in-class education. In the current scenario, efforts have been made to improve the communication between the instructors and the students by offering interactive forums, feedback forms, quiz etc. But there are still some problems that seperate online courses from in-class education. One of the main problems is detecting students confusion  level.  Unlike  in-class education,  where  a  teacher  can  judge  if  the  students  understand  the  materials  by  verbal  inquiries  or  their  body  language,  immediate feedback  from  the  student  is  less accessible  in  long  distance  education.  



## Dataset Information 

The dataset consists of EEG signal data collected from 10 college students while they watched MOOC video clips. The EEG signal can help detecting students mental states by acquiring signals from the brain.These 10 students were assigned to watch 20 videos, 10 of which were pre-labeled as “easy” and 10 as “difficult”. First, several online education videos were selected that were assumed not to be confusing for the students. Each video was about 2 minutes long. For “difficult” videos, the two-minute clip was taken abruptly from the middle of a topic to make the videos more confusing. The dataset was prepared by randomly selecting 5 videos from each category for each student. The dataset has a total of 12811 samples. On an average there are 120 samples for each data points. There are 100 datapoints, 10 coming from each student.

There are two label columns — user-defined label (self-labeled by the students based on their experience) and predefined label (where they are expected to be confused).

### Cleaning the data

```python
dataset = data.copy()
dataset = dataset[dataset['Attention'] > 0.0]
dataset = dataset[dataset['Mediation'] > 0.0]
dataset_all = dataset.copy()
dataset.drop(columns = ['SubjectID','VideoID','predefinedlabel'],inplace=True)
```

```python
print("After cleaning, the data has {} rows and {} columns.".format(dataset.shape[0],dataset.shape[1]))
```

### Normalization and standarization

```python
dataset.hist(figsize = (15,15), grid = False, color = "green", edgecolor = 'black',linewidth = 1.5 )
plt.suptitle('Histogram of different features in the dataset',fontsize = 20,y= 0.92)
plt.show()
```

**Comments**:<br>
The above histograms shows that most of the features have their distributions skewed. The histogram of target user-definedlabeln shows that the dataset is fairly balanced. The skewness in the distribution has to be be removed.Some of the features are not spread across the entire range. Also, feature scales are very different. So appropriate normalization needs to be applied to spread the data over the normalized range and yield maximum information from the given data.

```python
#Applying non-linear normalization for spreading the data
X_n = dataset.iloc[:,:-1].copy() #data that will be normalized
X_n.iloc[:,3:11] = np.log2(X_n.iloc[:,3:11] + 0.0001) #spreading the amount data using log normalization technique
X_n.iloc[:,3:11].hist(figsize = (15,15), grid = False, color = "green", edgecolor = 'black',linewidth = 1.5 )
```

**Comments**:<br>
So we managed to spread some of the features in the dataset. Let us now use a combination of tanh normalization and robust scaleing for transforming the data i.e. spreading the data and scaling it in the range [-1,1]

```python
m = np.mean(X_n, axis=0) # array([16.25, 26.25])
std = np.std(X_n, axis=0) # array([17.45530005, 22.18529919])
md = np.median(X_n,axis = 0)
p75 = np.percentile(X_n,75,axis = 0)
p25 = np.percentile(X_n,25,axis = 0)
```

```python
X_n = np.tanh((X_n - md) / (p75 - p25)) #normalization and feature scaling for train set
```

```python
X_n.hist(figsize = (15,15), grid = False, color = "green", edgecolor = 'black',linewidth = 1.5 )
plt.suptitle('Histogram of different features in the dataset after normalization',fontsize = 20,y= 0.92)
plt.show()
```

**Comment**: <br>
All the features have been spread and scaled in the range [-1,1].

```python
plt.figure(figsize = (15,15))
cor_matrix = pd.concat((X_n,dataset_all['user-definedlabeln']),axis = 1).corr()
sns.heatmap(cor_matrix,annot=True)
```

# **3. Creating datasets**

```python
X = X_n #features
Y = dataset.iloc[:,-1] #targets
```

```python
#identify the minimum and maximum length of time-series data in the dataset
VideoID = list(set(data.iloc[:,1]))
SubjectID = list(set(data.iloc[:,0])) 
Amin=150 # length of signal
Amax = 0
for i in range(len(SubjectID)):
    for j in range(len(VideoID)):
        Xtemp=data[(data.iloc[:,0]==SubjectID[i]) & (data.iloc[:,1]==VideoID[j])]
        Amin = min(len(Xtemp.iloc[:,14]),Amin)
        Amax = max(len(Xtemp.iloc[:,14]),Amax)
print("The minimum length of the time series data is {}.".format(Amin))
print("The maximum length of the time series data is {}.".format(Amax))
```

#### Creating dataset with minimum length of the time series.

```python
X = np.asarray(X)
Y = np.asarray(Y)
S = np.asarray(dataset_all.iloc[:,0])
V = np.asarray(dataset_all.iloc[:,1])
X1 = {}
Y1 = {}
for i in range(X.shape[0]): #for each row of features
    s = int(S[i]*10 + V[i])
    if s not in X1.keys():
        X1[s] = X[i]
    elif X1[s].shape[0]<Amin*11: # 112*11
        X1[s] = np.concatenate((X1[s],X[i]),axis =0)
    Y1[s]= Y[i]
```

```python
length_dict = {key: len(value) for key, value in X1.items()}
len(set(length_dict.values())) <= 1 #all values are equal
```

#### Creating dataset with maximum length of the time series.

```python
X2 = {}
Y2 = {}
for i in range(X.shape[0]): #for each row of features
    s = int(S[i]*10 + V[i])
    if s not in X2.keys():
        X2[s] = X[i]
    else:
        X2[s] = np.concatenate((X2[s],X[i]),axis =0)
    Y2[s]= Y[i]
    
for key in X2.keys():
    z = np.zeros(Amax*11 - len(X2[key])) #144X11
    X2[key] = np.concatenate((X2[key],z),axis = 0) #zero padding at the end
```

```python
length_dict = {key: len(value) for key, value in X2.items()}
len(set(length_dict.values())) <= 1 #all values are equal
```

```python
D1 = np.zeros((100,Amin*11),dtype = float)
T = np.zeros((100,1),dtype = int)
D2 = np.zeros((100,Amax*11),dtype = float)
for i in X1.keys():
    D1[i,:] = X1[i]
    D2[i,:] = X2[i]
    T[i] = Y1[i]
```

```python
D1 = D1.reshape(-1,1,Amin,11,1)
D2 = D2.reshape(-1,1,Amax,11,1)
```


# **4.Predictive modelling using CNN + Bi-Directional LSTM**

Using CNN along with LSTM networks helps extracting useful signal features from the sequences.

```python
def plotLearningCurve(history,epochs,text):
  epochRange = range(1,epochs+1)
  plt.figure(figsize = (5,5))
  plt.plot(epochRange,history.history['accuracy'])
  plt.plot(epochRange,history.history['val_accuracy'])
  plt.title('Model Accuracy for ' + text)
  plt.xlabel('Epoch', fontsize = 20)
  plt.ylabel('Accuracy', fontsize = 20)
  plt.legend(['Training set','Validation set'])
  plt.show()
  
  plt.figure(figsize = (5,5))
  plt.plot(epochRange,history.history['loss'])
  plt.plot(epochRange,history.history['val_loss'])
  plt.title('Model Loss for ' + text)
  plt.xlabel('Epoch', fontsize = 20)
  plt.ylabel('Loss', fontsize = 20)
  plt.legend(['Training set','Validation set'])
  plt.show()
```

```python
def lstm_model(Xtrain,Xtest,ytrain,ytest,maxlen,featlen,epochs):
    i = Input(shape=(None,maxlen,featlen,1)) #maxlen gives us the number of temporal slices per data point
    x = TimeDistributed(Conv2D(16,(3,3),activation = 'tanh'))(i)
    x = Dropout(0.05)(x)
    x = TimeDistributed(Conv2D(32,(5,5),activation = 'tanh'))(x)
    x = Dropout(0.05)(x)
    x = TimeDistributed(MaxPooling2D(pool_size=(2, 2)))(x)
    x = TimeDistributed(Flatten())(x)
    x = Bidirectional(LSTM(75, return_sequences=True),merge_mode = 'ave')(x)
    x = Dense(1, activation='sigmoid')(x)
    model = Model(i, x)
    model.summary()
    #Training the LSTM
    model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
    r = model.fit(Xtrain,ytrain, 
                  validation_data = (Xtest,ytest), 
                  epochs = epochs, 
                  verbose = 2,
                  batch_size = 32,
                 shuffle = True)
    print("Train score:", model.evaluate(Xtrain,ytrain))
    print("Test score:", model.evaluate(Xtest,ytest))
    return r,model
```

```python
#For dataset D1
Xtrain, Xtest, ytrain, ytest = train_test_split(D1, T, test_size=0.2, random_state=0)
r,model = lstm_model(Xtrain,Xtest,ytrain,ytest,Amin,11,epochs = 100)
```

```python
plot_model(model)
```

```python
 plotLearningCurve(r,100,text = 'Dataset 1')
```

```python
#Generate predictions for the test dataset
ypred = model.predict(Xtest,batch_size = 32).flatten()
ypred = ypred>0.5
#Get the confusion matrix
cf_matrix = confusion_matrix(ytest, ypred)
sns.heatmap(cf_matrix,annot = True,fmt ='g', cmap='Blues')
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('Confusion Matrix for Dataset 1')
plt.show()
```

```python
#For dataset D2
Xtrain, Xtest, ytrain, ytest = train_test_split(D2, T, test_size=0.2,random_state = 0)
r,model = lstm_model(Xtrain,Xtest,ytrain,ytest,Amax,11,epochs = 100)
```

```python
 plotLearningCurve(r,100,text = 'Dataset 2')
```
