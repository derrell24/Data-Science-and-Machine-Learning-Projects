# X-Ray Image Classification for COVID-19 and Pneumonia - A Deep Learning Approach using Tensorflow

In this notebook we will try to address the problem of classifying normal images, covid -19 images and pneumonia images

### Dataset Information

In our analysis we will be using the following dataset
* **COVID-19 RADIOGRAPHY DATABASE** \
A team of researchers from Qatar University, Doha, Qatar, and the University of Dhaka, Bangladesh along with their collaborators from Pakistan and Malaysia in collaboration with medical doctors have created a database of chest X-ray images for COVID-19 positive cases along with Normal and Viral Pneumonia images. The database consists of 3616 COVID-19 positive cases along with 10,192 Normal, 6012 Lung Opacity (Non-COVID lung infection), and 1345 Viral Pneumonia images.[dataset](https://www.kaggle.com/tawsifurrahman/covid19-radiography-database) 

### Importing Libraries

```python
import os 
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory
import seaborn as sns
import numpy as np
```

### Creating dataset

```python
#create training and testing datasets
train_data = image_dataset_from_directory(
    './COVID-19_Radiography_Dataset',
    labels="inferred",
    label_mode="int",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(244, 244),
    shuffle=True,
    seed=123,
    validation_split=0.2,
    subset="training",
    interpolation="bilinear",
    follow_links=False,
)
```

```python
#create training and testing datasets
test_data = image_dataset_from_directory(
    './COVID-19_Radiography_Dataset',
    labels="inferred",
    label_mode="int",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(244, 244),
    shuffle=True,
    seed=123,
    validation_split=0.2,
    subset="validation",
    interpolation="bilinear",
    follow_links=False,
)
```

```python
class_names = train_data.class_names
print(class_names)
```

```python
#visualize data
plt.figure(figsize=(10, 10))
for images, labels in train_data.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```

```python
label_array_train = np.array([])
for images, labels in train_data.take(-1):  # only take first element of dataset, -1 takes all the elements of the dataset
    #numpy_images = images.numpy()
    numpy_labels =  labels.numpy()
    label_array_train = np.append(label_array_train,numpy_labels)
```

```python
ax = sns.countplot(label_array_train,label = class_names)
ax.set_xticklabels(class_names)
ax.set_title('Training data class counts')
```

```python
label_array_test = np.array([])
for images, labels in test_data.take(-1):  # only take first element of dataset, -1 takes all the elements of the dataset
    #numpy_images = images.numpy()
    numpy_labels =  labels.numpy()
    label_array_test = np.append(label_array_test,numpy_labels)
```

```python
ax = sns.countplot(label_array_test,label = class_names)
ax.set_xticklabels(class_names)
ax.set_title('Testing data class counts')
```
